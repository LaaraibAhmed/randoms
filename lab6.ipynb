```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

filepath="/content/lab05.tif"
im=Image.open(filepath)
im_array=np.array(im)
im_array

```

```python
fig = plt.figure(figsize=(10, 10))
rows=2
columns=4
def plot(x,figure):
  fig.add_subplot(rows, columns, x)
  plt.imshow(figure,cmap="gray")
```

```python
# for x in range(1,7):
plane=np.left_shift(im_array,1)
plane=np.right_shift(plane,1)
plot(1,plane)
plane=np.left_shift(im_array,2)
plane=np.right_shift(plane,2)
plot(2,plane)
```
